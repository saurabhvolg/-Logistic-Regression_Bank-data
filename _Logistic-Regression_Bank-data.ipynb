```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
```

```python
bank = pd.read_csv('/Users/SAURABH/Saurabh patil/DATA SCIENCE/LOgistic Regression/bank-full.csv',sep=';')
```

```python
#Checking all the columns
bank.head()
```

```python
bank.y.replace(('yes','no'),(1,0),inplace=True)
```

```python
bank.default.replace(('yes','no'),(1,0),inplace=True)
```

```python
bank.housing.replace(('yes','no'),(1,0),inplace=True)
```

```python
bank.loan.replace(('yes','no'),(1,0),inplace=True)
```

```python
bank.head(3)
```

```python
#Checking for na values
bank.info()
```

```python
bank = pd.get_dummies(bank)
```

```python
bank.head(3)
```

```python
bank.info()
```

```python
bank[bank.duplicated()]
```

```python
#Dividing the dataset into X and Y variables
X = bank.loc[:,bank.columns!='y']
Y = np.ravel(bank.loc[:,bank.columns=='y'])
```

```python
X
```

```python
Y
```

```python
#Building the logistic regression modellalit mehendiratta
model = LogisticRegression()
model.fit(X,Y)
```

```python
#Predict for X dataset
y_pred = model.predict(X)
```

```python
y_pred_df= pd.DataFrame({'actual': Y,
                         'predicted_prob': y_pred})
```

```python
y_pred_df
```

```python
# Confusion Matrix for the model accuracy
from sklearn.metrics import confusion_matrix
confusion_matrix = confusion_matrix(Y,y_pred)
print (confusion_matrix)
```

```python
#Classification report
from sklearn.metrics import classification_report
print(classification_report(Y,y_pred))
```

```python
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score

fpr, tpr, thresholds = roc_curve(Y, model.predict_proba (X)[:,1])

auc = roc_auc_score(Y, y_pred)

import matplotlib.pyplot as plt
plt.plot(fpr, tpr, color='red', label='logit model ( area  = %0.2f)'%auc)
plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel('False Positive Rate or [1 - True Negative Rate]')
plt.ylabel('True Positive Rate')
```
